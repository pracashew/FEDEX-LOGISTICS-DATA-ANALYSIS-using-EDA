````python
# Chart - 1 visualization code
# lets find outliers
# to find ouliers we use boxplot ,boxplot is best all the time
sns.boxplot(data=df,x="Line Item Value")
plt.show()
# in boxplot we can identify directions of outliers
# outliers are seen towards max prices
````

````python
Q1=df.loc[:,"Line Item Value"].quantile(0.25)
Q3=df.loc[:,"Line Item Value"].quantile(0.75)
IQR=Q3-Q1
print("Q1:",Q1)
print("Q3:",Q3)
print("IQR:",IQR)
````

````python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
print("lower_limit:",lower_limit)
print("upper_limit:",upper_limit)
````

````python
#lets filter a data
# lets see is there any negative outliers
df[df.loc[:,"Line Item Value"]<lower_limit]
````

````python
#lets filter a data
# lets see is there any pos outliers
df[df.loc[:,"Line Item Value"]>upper_limit]
````

````python
#lets see whats the percentage of outliers ,do we have more or less
len(df[df.loc[:,"Line Item Value"]>upper_limit])/len(df)*100
````

````python
#lets drop outliers since % is very less
#identify rows where the price is outside the specified range
outliers=df[(df["Line Item Value"]<lower_limit)|(df["Line Item Value"]>upper_limit)].index
df=df.drop(outliers)
# .index get incates of row
# drop the rows
````

````python
#resent the index after dropping rows
df.reset_index(drop=True,inplace=True)
df
````
